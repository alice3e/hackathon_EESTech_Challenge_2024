## Imports

```python
import pandas as pd
import numpy as np
```

## Loading data

```python
PATH_TO_DATA = '../data/для анализа/'
```

```python
# Loading data
df_anomaly = pd.read_csv(PATH_TO_DATA + 'dataset._anomaly.csv', sep=';')
df_normal = pd.read_csv(PATH_TO_DATA + 'dataset._normal.csv', sep=';')
df_problems = pd.read_csv(PATH_TO_DATA + 'dataset._problems.csv', sep=';')
```

## Cleaning data

```python
# Смотрим на процентное количество пропущенных значений
print(f'DF: ANOMALY\n{df_anomaly.isna().sum() / len(df_anomaly)}')

# Не целесообразно оставлять признаки без значений
df_anomaly.drop(columns=['iButton2'], axis=1, inplace=True)
```

```python
# Аналогично для остальных датасетов
print(f'DF: NORMAL\n{df_normal.isna().sum() / len(df_normal)}\n')
 
df_normal.drop(columns=['iButton2'], axis=1, inplace=True)

print(f'DF: PROBLEMS\n{df_problems.isna().sum() / len(df_problems)}')
 
df_problems.drop(columns=['iButton2'], axis=1, inplace=True)
```

```python
# Смотрим, что у датасетов одинаковое количество признаков
print(f'Количество признаков в датасетах: {len(df_problems.columns), len(df_anomaly.columns), len(df_normal.columns)}')

# Проверяем, что признаки одинаковые
print(f'Количество совпадающих признаков: {len(set(df_problems.columns).intersection(set(df_anomaly.columns).intersection(set(df_normal.columns))))}')
```

```python
# Переводим даты в нужный формат
df_normal['Дата и время'] = pd.to_datetime(df_normal['Дата и время'], format='mixed')
df_anomaly['Дата и время'] = pd.to_datetime(df_anomaly['Дата и время'], format='mixed')
df_problems['Дата и время'] = pd.to_datetime(df_normal['Дата и время'], format='mixed')

# Генерируем признаки
def handle_datetime(df: pd.DataFrame) -> pd.DataFrame:
    df['Год'] = df['Дата и время'].dt.year
    df['Месяц'] = df['Дата и время'].dt.month
    df['День недели'] = df['Дата и время'].dt.day_of_week
    df['День недели'] = df['Дата и время'].dt.quarter

    df.drop(columns=['Дата и время'], axis=1, inplace=True)

    return df

df_normal = handle_datetime(df_normal)
df_anomaly = handle_datetime(df_anomaly)
df_problems = handle_datetime(df_problems)
```

```python
# Признак нагрузки на двигатель не информативен, так как принимает всего 1 значение

df_anomaly.drop(columns=['Нагрузка на двигатель, %'], axis=1, inplace=True)
df_normal.drop(columns=['Нагрузка на двигатель, %'], axis=1, inplace=True)
df_problems.drop(columns=['Нагрузка на двигатель, %'], axis=1, inplace=True)
```

```python
# Функция для вычисления неинформативных колонок
def uninformative_columns(df: pd.DataFrame) -> list:
    result = []
    for col in df.columns:
        if df[col].nunique() < 2:
            result.append(col)
    return result

cols_to_drop = set(uninformative_columns(df_problems)).intersection(set(uninformative_columns(df_normal)).intersection(set(uninformative_columns(df_anomaly))))
```

```python
# Неинформативные колонки
cols_to_drop
```

```python
# Удаляем их
df_normal.drop(columns=cols_to_drop, axis=1, inplace=True)
df_problems.drop(columns=cols_to_drop, axis=1, inplace=True)
df_anomaly.drop(columns=cols_to_drop, axis=1, inplace=True)
```

```python
# Категориальные признаки
cat_features = [
    'Сост.пед.сцепл.',
]

# Бинарные признаки
bin_features = [
    'Нейтраль КПП (spn3843)',
    'Стояночный тормоз (spn3842)',
    'Аварийная температура охлаждающей жидкости (spn3841)',
    'Засоренность воздушного фильтра (spn3840)',
    'Засоренность фильтра КПП (spn3847)',
    'Аварийное давление масла ДВС (spn3846)',
    'Засоренность фильтра ДВС (spn3845)',
    'Засоренность фильтра рулевого управления (spn3844)',
    'Засоренность фильтра навесного оборудования (spn3851)',
    'Недопустимый уровень масла в гидробаке (spn3850)',
    'Аварийная температура масла в гидросистеме (spn3849)',
    'Аварийное давление в I контуре тормозной системы (spn3848)',
    'Аварийное давление в II контуре тормозной системы (spn3855)',
    'Зарядка АКБ (spn3854)',
    'Отопитель (spn3853)',
    'Выход блока управления двигателем (spn3852)',
    'Включение тормозков (spn3859)',
    'Засоренность фильтра слива (spn3858)',
    'Аварийное давление масла КПП (spn3857)',
    'Аварийная температура масла ДВС(spn3856)',
    'Неисправность тормозной системы (spn3863)',
    'Термостарт (spn3862)',
    'Разрешение запуска двигателя (spn3861)',
    'Низкий уровень ОЖ (spn3860)',
    'Аварийная температура масла ГТР (spn3867)',
    'Необходимость сервисного обслуживания (spn3866)',
    'Подогрев топливного фильтра (spn3865)',
    'Вода в топливе (spn3864)',
    'Холодный старт (spn3871)',
]

# Числовые признаки
num_features = [
    'Значение счетчика моточасов, час:мин',
    'Давление в пневмостистеме (spn46), кПа',
    'ДВС. Температура охлаждающей жидкости',
    'ДВС. Давление смазки',
    'КПП. Давление масла в системе смазки',
    'Давл.масла двиг.,кПа',
    'ДВС. Частота вращения коленчатого вала',
    'Электросистема. Напряжение',
    'Уровень топлива % (spn96)',
    'Скорость',
    'КПП. Температура масла',
    'Обор.двиг.,об/мин',
    'Полож.пед.акселер.,%',
    'Месяц',
    'День недели',
    'Темп.масла двиг.,°С'
]
```

```python
# Обработка категориального признака
df_normal['Сост.пед.сцепл.'] = df_normal['Сост.пед.сцепл.'].replace('        -', np.nan)
df_problems['Сост.пед.сцепл.'] = df_problems['Сост.пед.сцепл.'].replace('        -', np.nan)
df_anomaly['Сост.пед.сцепл.'] = df_anomaly['Сост.пед.сцепл.'].replace('        -', np.nan)

# Обработка бинарных признаков
def list_bin_cols(df: pd.DataFrame) -> pd.DataFrame:
    for col in bin_features:
        df[col].replace('-', '', inplace=True)
        df[col].fillna('', inplace=True)
        df[col] = df[col].astype(str)
        if df[col].nunique() > 2:
            print(col)
    
    return df
    

df_normal = list_bin_cols(df_normal)
df_problems = list_bin_cols(df_problems)
df_anomaly = list_bin_cols(df_anomaly)
```

```python
# Обработка числовых признаков
def handle_num_cols(df: pd.DataFrame) -> pd.DataFrame:
    for col in num_features:
        df[col].replace('-', np.nan, inplace=True)
        df[col].replace('        -', np.nan, inplace=True)
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(-1).astype(int)
    
    return df


df_normal = handle_num_cols(df_normal)
df_problems = handle_num_cols(df_problems)
df_anomaly = handle_num_cols(df_anomaly)
```

```python
# Добавляем отсутствия наблюдения как признак
def create_is_na_columns(df: pd.DataFrame) -> pd.DataFrame:
    for column in df.columns:
        new_column_name = f"{column}_is_na"
        df[new_column_name] = df[column].isna().astype(int)
    
    return df

df_normal = create_is_na_columns(df_normal)
df_problems = create_is_na_columns(df_problems)
df_anomaly = create_is_na_columns(df_anomaly)
```
